```python
#q2
from abc import ABC, abstractmethod

class shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Triangle(shape):
    def __init__(self,b,h):
        self.b=b
        self.h=h
    def area(self):
        return 0.5*self.b*self.h
class Rectangle(shape):
    def __init__(self,l,w):
        self.l=l
        self.w=w
    def area(self):
        return 0.5*self.l*self.w
class Square(shape):
    def __init__(self,l):
        self.l=l
    def area(self):
        return 0.5*self.l*self.l
   

t=Triangle(3,4)
r=Rectangle(3,5)
s=Square(4)

print("Triangle are ",t.area())
print("Rectangle area ",r.area())
print("Square area ",s.area())
```

```python
#q3
class Student:
    def __init__(self,id,name):
        self.id = id
        self.name = name

    def student_details(self):
        print("Student ID ",self.id)
        print("Student Name ",self.name)

class Marks(Student):
    def __init__(self,id,name,marks_algo,marks_dataScience,marks_calculus):
        super().__init__(id,name)
        self.marks_algo = marks_algo
        self.marks_dataScience = marks_dataScience
        self.marks_calculus = marks_calculus

    def display_marks(self):
        print("Marks in Algorithms ", self.marks_algo)
        print("Marks in Data Science ", self.marks_dataScience)
        print("Marks in Calculus ", self.marks_calculus)

class Result(Marks):
    def __init__(self,id,name,marks_algo,marks_dataScience,marks_calculus):
        super().__init__(id,name, marks_algo,marks_dataScience,marks_calculus)
        self.total_marks = self.marks_algo+self.marks_dataScience+self.marks_calculus
        self.average_marks = self.total_marks / 3

    def display_result(self):
        print("Total Marks ", self.total_marks)
        print("Average Marks ", self.average_marks)

result = Result(4005,"Mujtaba Saqib",67,90,85)
result.student_details()
result.display_marks()
result.display_result()
```

```python
#q5
class Vehicle:
    def __init__(self, make, model, rental_price):
        self._make = make
        self._model = model
        self._rental_price = rental_price
        self._availability_status = "available"

    def is_available(self):
        return self._availability_status == "available"

    def calculate_total_rental_cost(self, rental_period):
        return self._rental_price * rental_period

    def display_vehicle_details(self):
        print(f"Make: {self._make}")
        print(f"Model: {self._model}")
        print(f"Rental Price: {self._rental_price}")
        print(f"Availability Status: {self._availability_status}")

    def rent(self):
        self._availability_status = "rented"

    def return_vehicle(self):
        self._availability_status = "available"


class Car(Vehicle):
    pass


class SUV(Vehicle):
    pass


class Truck(Vehicle):
    pass


class RentalReservation:
    def __init__(self, customer, vehicle, rental_start_date, rental_end_date):
        self.customer = customer
        self.vehicle = vehicle
        self.rental_start_date = rental_start_date
        self.rental_end_date = rental_end_date

    def display_rental_reservation_details(self):
        print("Customer {} ".format(self.customer.name))
        print("Vehicle {} {}".format(self.vehicle._make,self.vehicle._model))
        print("Rental Start Date {}".format(self.rental_start_date))
        print("Rental End Date {}".format(self.rental_end_date))


class Customer:
    def __init__(self, name, contact_information):
        self.name = name
        self.contact_information = contact_information
        self._rented_vehicles = []

    def rent_vehicle(self, vehicle):
        self._rented_vehicles.append(vehicle)
        vehicle.rent()

    def return_vehicle(self, vehicle):
        self._rented_vehicles.remove(vehicle)
        vehicle.return_vehicle()

    def display_rental_history(self):
        print("Customer Rental History for {}".format(self.name))
        for vehicle in self._rented_vehicles:
            print("{} {}".format(vehicle._make,vehicle._model))


def display_details(obj):
    if isinstance(obj, Vehicle):
        obj.display_vehicle_details()
    elif isinstance(obj, RentalReservation):
        obj.display_rental_reservation_details()
    else:
        raise TypeError("Invalid object type")

car = Car("Toyota","GLI",60)
is_available = car.is_available()
print("Is the car available? {}".format(is_available))
total_rental_cost = car.calculate_total_rental_cost(3)
print("Total rental cost for 3 days {}".format(total_rental_cost))
display_details(car)

customer = Customer("Muji","muji@gailcom")
customer.rent_vehicle(car)

rental_reservation = RentalReservation(customer,car,"2023-10-01","2023-10-04")
display_details(rental_reservation)

customer.return_vehicle(car)
customer.display_rental_history()
```

```python
#q6
class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary

    def calculate_bonus(self):
        bonus_amount = self.salary * 0.1

        if isinstance(self,Manager):
            bonus_amount = self.salary*0.2
        elif isinstance(self,SeniorManager):
            bonus_amount = self.salary*0.3

        return bonus_amount

class Manager(Employee):
    def hire(self):
        print("{} is hiring someone".format(self.name))

class Developer(Employee):
    def write_code(self):
        print("{} is writing code".format(self.name))

class SeniorManager(Manager):
    def calculate_bonus(self):
        bonus_amount = self.salary*0.3
        return bonus_amount

manager = Manager("Arsal",100000)
developer = Developer("Sheri",65000)
senior_manager = SeniorManager("Mujtaba",190000)

manager_bonus = manager.calculate_bonus()
developer_bonus = developer.calculate_bonus()
senior_manager_bonus = senior_manager.calculate_bonus()

print("Manager bonus ", manager_bonus)
print("Developer bonus ", developer_bonus)
print("Senior manager bonus ", senior_manager_bonus)

manager.hire()
developer.write_code()
```
