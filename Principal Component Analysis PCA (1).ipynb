```python
pca_30=PCA(n_components=30)
pca_30.fit(scaled_X)
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_varaince=[]
for n in range(1,30):
    pca=PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_varaince.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_varaince,color='darkblue')
plt.xlabel("Number of components")
plt.ylabel("Variance Explained")
plt.grid(True)
plt.show()
```
