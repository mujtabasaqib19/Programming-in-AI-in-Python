```python
#q2
import pandas as pd
df = pd.read_csv('C:\\Users\\mujta\\Downloads\\iris.data')
names_of_coloumns = ['sepal_length','sepal_width','petal_length','petal_width','class']
df.columns = names_of_coloumns
print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

array = df.values
X = array[:, 0:4]
y = array[:, 4]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

a_scores = []
for k in range(1, min(251, len(X_train) + 1)):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_train)
    accuracy = accuracy_score(y_train, y_pred)
    a_scores.append(accuracy)

plt.figure(figsize=(10, 6))
plt.plot(range(1, min(251, len(X_train) + 1)), a_scores, marker='o', linestyle=':', color='b')
plt.title('Accuracy vs. Number of Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
max=a_scores[0]
for j in a_scores:
  if(j>max):
    max=j
    k=k+1
print("Neighbour {} highest accuracy {}".format(k,max))
```

```python
min=a_scores[0]
for j in a_scores:
  if(j<min):
    min=j
    k=k+1
print("Neighbour {} lowest accuracy {}".format(k,min))
```
